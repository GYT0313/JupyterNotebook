```python
import numpy as np

# 1、加载数据，保存到数组中
dtype = np.dtype([('Site', 'S20'), ('Variety', 'S20'), ('Yield', 'f4'), ('Year', 'i4')])
data = np.loadtxt('./barley2.csv', dtype=dtype, delimiter=',', skiprows=1)
site, variety, yyield, year = np.loadtxt('./barley2.csv', 
                                         dtype=dtype, 
                                         delimiter=',', 
                                         skiprows=1, 
                                         unpack=True)
print(data)
```

```python
# 2、按地点将数组进行分割
# 先排序
sort_by_site = np.sort(data, kind='mergesort', order='Site')
# 再改变纬度
sort_by_site.shape = 120,1
# 再分组
print(np.vsplit(sort_by_site, len(np.unique(site))))
```

```python
# 3、按小麦品种将数组进行分割
# 先排序
sort_by_site = np.sort(data, kind='mergesort', order='Variety')
# 再改变纬度
sort_by_site.shape = 120,1
# 再分组
print(np.vsplit(sort_by_site, len(np.unique(variety))))
```
