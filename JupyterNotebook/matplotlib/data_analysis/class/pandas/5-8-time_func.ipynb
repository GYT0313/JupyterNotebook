```python
# 通过年、月、日构建
ad = {"Year":[2018,2019], "Month":[5,6], "Day":[9,10]}
df = pd.DataFrame(ad)
print(df)
t5 = pd.to_datetime(df)
print(t5)
df["new_date"] = t5
print(df)
```

```python
y = [x.year for x in df["new_date"]]
print(y)
m = [x.month for x in df["new_date"]]
print(m)
d = [x.day for x in df["new_date"]]
print(d)
```

```python
# set_index
df2 = df.set_index("new_date")
print(df2)
print(df)

print(df2.index)

# reset_index
df3 = df2.reset_index()
print(df3)

# 增加一行DateTime 索引
t6 = pd.DatetimeIndex(t3)
print(t6)

```
