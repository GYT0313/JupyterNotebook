```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("./stu_scores.csv")
# 分组
group_df = data.groupby("Gender")
list_df = list(group_df)
print(list_df[0][1])
print(group_df.mean())
```

```python
print(data.head(3))
# 聚合 agg
agg_data = data["Chinese"].agg(lambda x: x+10)
print(agg_data.head(3))

agg2_data = data.agg({"Chinese": lambda x: x+10, "Math": lambda x: x+5})
print(agg2_data.head(3))

agg3_data = data["Chinese"].agg([np.sum, np.max, np.min])
print(agg3_data.head(3))

# 分组聚合
agg4_data = group_df.agg([np.sum, np.max, np.min])
agg4_data.head(3)
```

```python
# apply
app1_data = group_df.apply(np.mean)
print(app1_data)

# transform
trans1_data = group_df.transform(pd.DataFrame.mean)
print(trans1_data.head(3))

print("="*10)

# apply
app2_data = data["Chinese"].apply(np.mean)
print(app1_data)

# transform
trans2_data = data["Chinese"].transform(lambda x: x+100)
print(trans2_data.head(3))

trans3_data = data["Chinese"].transform(lambda x: np.max(x))
print(trans3_data.head(3))
```

```python
# 透视表
pivot_data = pd.pivot_table(data, values=["Chinese", "Math"], index="Gender", aggfunc=[np.sum, np.max, np.min, np.mean])
pivot_data
```
